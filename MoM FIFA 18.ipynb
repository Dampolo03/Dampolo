```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('FIFA 2018 Statistics.csv')
```

```python
data.head()
```

```python
data_copy = data.copy()
```

```python
data_copy['Man of the Match'].replace(['Yes', 'No'], [1, 0], inplace = True)
data_copy['PSO'].replace(['Yes', 'No'], [1, 0], inplace = True)
```

```python
data_copy.fillna(0, inplace = True)
```

```python
data_copy = data_copy.drop(['Date', 'Opponent', 'Round'], axis = 1)
```

```python
def one_hot_encoding(data, col_list):
    return pd.get_dummies(data, columns = col_list)
```

```python
data_copy = one_hot_encoding(data_copy, ['Team'])
```

```python
data_copy.head()
```

```python
train, test = train_test_split(data_copy, test_size = 0.2, random_state = 42)
```

```python
y_train = train['Man of the Match']
x_train = train.drop(['Man of the Match'], axis = 1)

y_test = test['Man of the Match']
x_test = test.drop(['Man of the Match'], axis = 1)
```

```python
train.head()
```

```python
model = xgb.XGBClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
predict_train = model.predict(x_train)
```

```python
accuracy_train = accuracy_score(y_train, predict_train)
```

```python
accuracy_train
```

```python
model.fit(x_test, y_test)
```

```python
predict_test = model.predict(x_test)
```

```python
accuracy_test = accuracy_score(y_test, predict_test)
```
