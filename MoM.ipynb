```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn.metrics import mean_squared_error, regression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
#from sklearn.tree import DecisionTreeRegressor
```

```python
def one_hot_encoding(data, col_list):
    return pd.get_dummies(data, columns = col_list)
```

```python
train = one_hot_encoding(train, ['Team'])
```

```python
train = train.drop(['Date', 'Opponent', 'Round'], axis = 1)
```

```python
train_copy = train.copy()
```

```python
y = train_copy['Man of the Match']
x = train_copy.drop(['Man of the Match'], axis = 1)
```

```python
lin_reg = LinearRegression()
lasso = Lasso(alpha = 0.05)
ridge = Ridge(alpha = 0.05)
svr = SVR()
knn = KNeighborsClassifier(n_neighbors=10)
#tree = DecisionTreeRegressor()
```

```python
y = y.astype(int)
```

```python
lin_reg.fit(x, y)
lasso.fit(x, y)
ridge.fit(x, y)
svr.fit(x, y)
#tree.fit(x, y)
```

```python
y_pred1 = lin_reg.predict(x)
y_pred2 = lasso.predict(x)
y_pred3 = ridge.predict(x)
y_pred4 = svr.predict(x)
```

```python
print(np.sqrt(mean_squared_error(y, y_pred1)))
print(np.sqrt(mean_squared_error(y, y_pred2)))
print(np.sqrt(mean_squared_error(y, y_pred3)))
print(np.sqrt(mean_squared_error(y, y_pred4)))
```

```python
rmse1 = np.sqrt(-(cross_val_score(lin_reg, x, y, scoring = 'neg_mean_squared_error', cv = 10).mean()))
rmse2 = np.sqrt(-(cross_val_score(lasso, x, y, scoring = 'neg_mean_squared_error', cv = 10).mean()))
rmse3 = np.sqrt(-(cross_val_score(ridge, x, y, scoring = 'neg_mean_squared_error', cv = 10).mean()))
rmse4 = np.sqrt(-(cross_val_score(svr, x, y, scoring = 'neg_mean_squared_error', cv = 10).mean()))
```

```python
print(rmse1)
print(rmse2)
print(rmse3)
print(rmse4)
```
